```python
#from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import HTML, display
from scipy.signal import butter, lfilter, filtfilt
```

#### Parâmetros do experimento

```python
# 2 conditions: eyes open (OA) and eyes closed (OF)
# 4 participants
# 3 trials
# time  = 40s
# freq= 100Hz


freq = 100; #hertz
time = 40;  #segundos
freqCorte = 5;

#freqNiquist = freq/2
#lowpass 4th ordem
```

#### Leitura dos dados

```python
fname = "./kistler/qsceda1.txt"
data = np.loadtxt(fname, skiprows=1)
cop = data[:,7]
#copy = data[:,8]
t = np.linspace(0, time, freq*time)
```

#### Filtragem dos dados

```python
# Butterworth filter
b, a = butter(4, freqCorte/(freq/2), btype = 'low')
copxFiltrado = filtfilt(b, a, data[:,7])
copyFiltrado = filtfilt(b, a, data[:,8])
```

```python
plt.plot(t,copyFiltrado)
```
