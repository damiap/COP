#### Exercise 3 - What is supervised and unsupervised learning in neural networks. Provide examples.

#### Aprendizado supervisado: possui dois conjuntos de valores, de entrada e de saída. Assim, esta técnica consiste de um problema de otimização dos parâmetros da rede neural, ou seja, os pesos sinápticos, para responderem às atividades pré e pós sinápticas e de um sinal de erro. Depende de agente externo e pode ser aplicado para a execução do movimento do corpo do indivíduo, por exemplo, quando os sinais de um estímulo sensorial aferente são enviados para o Sistema Nervoso Central, e este envia respostas motoras adequadas para que a pessoa desempenhe um determinado movimento.

#### Aprendizado não supervisado: faz ajuste dos pesos através de valores de entrada e utiliza o conceito de atrator. Não depende de agente externo e pode ser aplicado em testes de memória associativa, por exemplo, fornecer figuras de diferentes objetos (maçã, flor, peixe, laranja, cachorro, gato, banana, árvore) e solicitar ao indivíduo para agrupar as figuras correspondentes, outro exemplo, sentir um aroma de um perfume e associá-lo com uma pessoa.

#### Exercise 4-  In signal processing it is usual to analyze the data in time and frequency domain. 

#### a) Explain one technique of analysis in the time domain and another in the frequency domain. 

#### Domínio Tempo: o período em que uma função se repete a cada intervalo de tempo, como T, 2T, 3T, 4T. É representado pela fórmula: T = 1/f. Analisar as wavelets dos sistemas olfativo, respitatório, cardíaco, por exemplo, são mais eficientes utilizando o domínio tempo.

#### Domínio Frequência: corresponde a frequência que um sinal ocorre em determinado intervalo de tempo. É representado pela fórmula: f = 1/T. Os sinais de eletroencéfalograma e de eletromiograma são exemplos de medidas quantitativas que geram uma resposta a partir de um estímulo e podem ser utilizados para análise do domínio frequência.

#### A técnica da Transformada de Fourier é utilizada para transformar funções de domínio tempo em domínio frequência ou o inverso.

#### b)  What is peristimulus time histogram (PSTH)? What is the advantage and disadvantage of this technique?

#### Peristimulus time histogram é uma técnica utilizada para comparar duas sequências de eventos, como de potenciais de ação, e determinar por meio de histogramas o tempo em que dois potenciais de ação ocorrem, por exemplo, determinar se o neurônio A dispara potenciais de ação antes, depois ou ao mesmo tempo que o neurônio B.

#### A vantagem desta técnica está em saber se um neurônio disparou antes ou depois do outro.

#### A desvantagem do PSTH está em eliminar o comportamento médio do disparo.
