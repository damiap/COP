### Definitions

```python
# 2 conditions: open eyes (oe) and closed eyes (ce)
# 4 participants
# 1 task: quiet standing (qs)
# 3 trials
# time  = 60 s
# freq= 100 Hz
# freqNiquist = freq/2
# lowpass 4th order

# Definitions:
group = ['adults (control)'] 
# files with 9 variables = ['Time (s)', 'Fx', 'Fy', 'Fz', 'Mx', 'My', 'Mz', 'COPx', 'COPy']
task = 'qs' 
subjects = ['cf', 'da', 'hr', 'kj', 'md', 'rf']
conditions = ['ce', 'oe']
trials = ['01', '02', '03']
extension = '.txt'

# Time:
time = 60;  # seconds

# File directory:
path = './data_kistler/'

# Filter Butterworth
order = 4; # order filter
freqCorte = 10; # Hz
freq = 100; # Hz
```

```python
def COP_Processing (path, freq, time, freqCorte):
    """Leitura dos dados de um arquivo, filtra  o copx e o copy
    Parameters
    ----------
    fileNames : nome do arquivo
    freq : frequência de aquisição
    time : tempo de aquisição dos dados
    freqCorte : frequência de corte para um  filtro Butterworth passa baixa"""
    
    
    data = np.loadtxt(path, skiprows=1)
    copx = data[:,7]
    copy = data[:,8]
            
           
    b, a = butter(order, freqCorte/(freq/2), btype = 'low')
    copx = filtfilt(b, a, data[:,7])
    copy = filtfilt(b, a, data[:,8])    
    
    #copx = detrend(copx, type='constant')   
    #copy = detrend(copy, type='constant')
    
        
    copx = copx*100
    copy = copy*100
    
    
    return copx, copy
```

```python
for subject in subjects:       
        #fileNames = path + task[0] + condition + subject + extension
        
    for c, condition in enumerate(conditions):   
        for t, trial in enumerate(trials):
            fileNames = path + task + condition + subject + trial + extension
            print (fileNames)
            try:
                data = np.loadtxt(fileNames, skiprows=1, unpack=False)      
            except:
                #pass
                print('File ' + condition + subject + trial + extension + ' does not exist.')
            
            copx, copy = COP_Processing(fileNames, freq, time, freqCorte)

            
            """
            Welch's method [1]_ computes an estimate of the power spectral density
            by dividing the data into overlapping segments, computing a modified
            periodogram for each segment and averaging the periodograms.


            For the default 'hanning' window an overlap of 50% is a reasonable trade off between 
            accurately estimating the signal power, while not over counting any of the data.

            Reference:
            https://github.com/demotu/BMC/blob/master/functions/psd.py
            """


            def psd(x, fs, window, nperseg, noverlap, nfft, detrend, show, ax, scales, xlim, units):

                from scipy import signal, integrate

                if not nperseg:

                    nperseg = np.ceil(len(x) / 2) #ceil -> aproxima para o maior e mais próximo valor inteiro
                f, P = signal.welch(x, fs, window, nperseg, noverlap, nfft, detrend) # Estimate power spectral density using Welch's method
                Area = integrate.cumtrapz(P, f, initial=0)
                Ptotal = Area[-1]
                mpf = integrate.trapz(f * P, f) / Ptotal # mean power frequency
                fmax = f[np.argmax(P)]

                # frequency percentiles
                inds = [0]
                Area = 100 * Area / Ptotal # + 10 * np.finfo(np.float).eps

                for i in range(1, 2001): # or (1, 101)?????? """
                    inds.append(np.argmax(Area[inds[-1]:] >= i) + inds[-1])
                fpcntile = f[inds]

                if show:
                    _plot(x, fs, f, P, mpf, fmax, fpcntile, scales, xlim, units, ax)

                return fpcntile, mpf, fmax, Ptotal, f, P

            def _plot(x, fs, f, P, mpf, fmax, fpcntile, scales, xlim, units, ax):
                """Plot results of the ellipse function, see its help."""

                try:
                    import matplotlib.pyplot as plt
                except ImportError:
                    print('matplotlib is not available.')
                else:
                    if ax is None:
                        fig, ax = plt.subplots(1, 1, figsize=(7, 5))
                    if scales.lower() == 'semilogy' or scales.lower() == 'loglog':
                        ax.set_yscale('log')
                    if scales.lower() == 'semilogx' or scales.lower() == 'loglog':
                        ax.set_xscale('log')

                    plt.plot(f, P, linewidth=2)
                    ylim = ax.get_ylim()
                    plt.plot([fmax, fmax], [np.max(P), np.max(P)], 'ro',
                        label='Fpeak = %.2f' % fmax)
                    plt.plot([fpcntile[50], fpcntile[50]], ylim, 'r', lw=1.5,
                        label='F50%% = %.2f' % fpcntile[50])
                    plt.plot([mpf, mpf], ylim, 'r--', lw=1.5,
                        label='Fmean = %.2f' % mpf)
                    plt.plot([fpcntile[95], fpcntile[95]], ylim, 'r-.', lw=2,
                        label='F95%% = %.2f' % fpcntile[95])
                    leg = ax.legend(loc='best', numpoints=1, framealpha=.5,
                    title='Frequencies [Hz]')
                    plt.setp(leg.get_title(), fontsize=12)
                    plt.xlabel('Frequency [$Hz$]', fontsize=12)
                    plt.ylabel('Magnitude [%s$^2/Hz$]' % units, fontsize=12)
                    plt.title('Power spectral density', fontsize=12)
                    if xlim:
                        ax.set_xlim(xlim)
                    ax.set_ylim(ylim)
                    plt.tight_layout()
                    plt.grid()
                    plt.show()

                    return ax

                # fpcntile[50] -> gives the median power frequency in Hz.
                # mpf -> mean power frequency.
                # fmax ->  maximum power frequency
                # Ptotal -> total power

            #print ('fpcntile =', fpcntile)
            
            # COPX
            fpcntile, mpf, fmax, Ptotal, f, P  = psd(copx, fs=freq, window='hanning', nperseg=None, noverlap=None, nfft=None,  detrend='constant', show=True, ax=None, scales='linear', xlim=[0,10], units='cm')

            # COPY
            #fpcntile, mpf, fmax, Ptotal, f, P  = psd(copy, fs=freq, window='hanning', nperseg=None, noverlap=None, nfft=None,  detrend='constant', show=True, ax=None, scales='linear', xlim=[0,10], units='cm')
        
        
```
