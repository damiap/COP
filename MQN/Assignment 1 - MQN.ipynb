#### Functions

#### Exercise 1 - Plot the graph of the functions:

##### a) f(x) = x^2 - 3x + 2 -> for f(x) = y

```python
x, y = sym.symbols('x y')
y = x**2 - 3*x + 2
y
plot(y, (x, -100, 100));
```

##### b) f(t) = -t^3  -> for f(t) = f

```python
t, f = sym.symbols('f t')
f = - t**3
f
plot(f, (t, -10, 10));
```

##### c) h(t) = t^3 - 2t^2  -> for h(t) = m

```python
t, m = sym.symbols('m t')
m = t**3 - 2*t**2
m
plot(m, (t, -10, 10));
```

##### d) f(x) = cos(2x  − π) -1  -> for f(x) = y

```python
x, y = sym.symbols('x y')
y = cos(2*x - pi) - 1
y
plot(y, (x, -pi, pi));
```

##### e) g(y) = exp(-y/10)  -> for g(y) = h

```python
import math

y, h = sym.symbols('h y')
h = math.e**-y/10
h
plot(h, (y, -5, 5));
```

##### f) h(a) = tan(a) + b, for b = 0  -> for h(a) = g

```python
import pylab as pl
import numpy as np

a = np.linspace(-np.pi, np.pi, 256, endpoint=True)
g = np.tan(a) + 0

pl.plot(a, g)
pl.show()
```

##### g) y(x) = tan(-x) + d, for d = 1  -> for y(x) = n

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
n = np.tan(-x) + 1

pl.plot(x, n)
pl.show()
```

##### h) f(x) = cos(x**2)

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y = np.cos(x**2)

pl.plot(x, y)
pl.show()
```

##### i)f(x) = cos(sin(x)) 

```python
x, y = sym.symbols('x y')
y = cos(sin(x))
y
plot(y, (x, -pi, pi));
```
