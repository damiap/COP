```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# muscle
# Program to simulate muscle contractions using 3-component Hill model
```

```python
# muscle properties for human Vastus group
Lslack = 0.223       # slack length of SEE (element elastic series)
Umax = 0.04          # strain in SEE is 4% at Fmax
Lceopt = 0.093       # optimal length of CE (contractile element)
width = 0.63*Lceopt  # maximum length change of CE
Fmax = 7400          # maximal isometric force
a = 0.25*Fmax        # force velocity parameter a
b = 0.25*10*Lceopt    # force velocity parameter b (Nigg and Herzog, p. 174-175)


# set initial condition for state variable Lce and initialize ODE (ordinary derivative equation) solver
Lce = 0.087          # this makes SEE just slack at t=0
t = 0                # start time
#tt = np.arange(0, 2.99, 0.001) # start time, duration of experiment, integration step size in seconds
tend = 2.99          # duration of experiment 
h = 0.001            # integration step size in seconds
i = 1                # step counter
data = np.zeros([tend/h, 6])   # space to store time and force results
```

```python
# The passive and parallel elastic element (PE), which behaves as a nonlinear spring 
# where no force is generated below a certain length (the slack length) 
# and force increases with the muscle elongation

def passive(Lm=3.34, kpe=5, Umax=0.04):
    
    """Thelen (2003) force of the muscle parallel element.
    Parameters
    ----------
    Lm: muscle fiber length
    Lm = Lm/Lceopt -> Lm = 0.31/0.093 -> 3.34
    kpe: exponential shape factor
    Umax: passive muscle strain due to maximum isometric force

    Returns
    -------
    Fpe  : force of the muscle parallel (passive) element"""
    
    
    if Lm < 1:
        Fpe = 0
    else:
        Fpe = (np.exp(kpe*(Lm-1)/Umax)-1)/(np.exp(kpe)-1)
    
        
    return Fpe
    
```

```python
# start simulation

while t < tend:
    # prescribed ramp shortening profile
    if t <= 1:
        Lm = 0.31              # initial muscle+tendon length
    if t > 1 and t < 2:
        Lm = 0.31-0.04*(t-1)   # ramp shortening at 4 cm/s
        
              
           
    # Calcule force in SEE from current SEE length

    Lsee = Lm - Lce  # length of SEE is total length minus CE length
    if Lsee < Lslack:
        F = 0       # SEE is slack, so no force
    else:
        F = Fmax*((Lsee - Lslack)/(Umax*Lslack))**2   # SEE force-length relationship 

    
    
    # Calculate isometric force at this Lce from CE force-length relationship

    #F0 = np.max([0, Fmax*(1-((Lce-Lceopt)/width)**2)])
    
    Fl = Fmax*(1 - ((Lce - Lceopt)/width)**2)  
    F0 = np.max([0, Fl])

    # Calculate CE velocity from Hill's equation
    if F > F0:
        disp = 'Error: program cannot do eccentric contractions'

    Lcedot = -b*(F0-F)/(F+a)     # note: velocity is negative for shortening!

       
        
    # Calcule force in PEE from current PEE length
    Fpe = passive(Lm=0.31, kpe=5, Umax=0.04)
    
    
    # do one forward Euler integration step
    
    Lce = Lce + (h*Lcedot)
    t = t + h
    #print(t)
    i = i + 1
    data[i:] = [t, Lce, Lsee, Lcedot, F, Fpe]
```

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8)) = plt.subplots(4,2, figsize=(12, 10))
ax1.plot(data[:,0], data[:,4], 'k', linewidth=3)
ax1.set_xlabel('Time[s]')
ax1.set_ylabel('Force CE [N]')
ax2.plot(data[:,0], data[:,4], 'b', linewidth=3)
ax2.set_xlabel('Time [s]')
ax2.set_ylabel('Force SEE [N]')
ax3.plot(data[:,0], data[:,3], 'r', linewidth=3)
ax3.set_xlabel('Time [s]')
ax3.set_ylabel('Velocidade [m/s]')
ax4.plot(data[:,0], data[:,2], 'g', linewidth=3)
ax4.set_xlabel('Time [s]')
ax4.set_ylabel('Length SEE[m]')
ax5.plot(data[:,0], data[:,1], 'm', linewidth=3)
ax5.set_xlabel('Time [s]')
ax5.set_ylabel('Length CE [m]')
ax6.plot(data[:,0], data[:,5], 'c', linewidth=3)
ax6.set_xlabel('Time [s]')
ax6.set_ylabel('Force PEE [N]')
ax7.plot(data[:,1]+data[:,2], data[:,5], 'r', label='Passive', linewidth=5)
ax7.plot(data[:,1]+data[:,2], data[:,4], 'go', label='Active', linewidth=3)
ax7.plot(data[:,1]+data[:,2], data[:,4]+data[:,5], 'c', label='Total', linewidth=2)
ax7.legend(loc='best')
ax7.set_xlabel('Length [m]')
ax7.set_ylabel('Force [N]')
ax8.plot(data[:,0], data[:,4], 'mo', label='Active', linewidth=3)
ax8.plot(data[:,0], data[:,5], 'b', label='Passive', linewidth=3)
ax8.plot(data[:,0], data[:,4]+data[:,5], 'c', label='Total', linewidth=2)
ax8.legend(loc='best')
ax8.set_xlabel('Time [s]')
ax8.set_ylabel('Force Ative+Passive [N]')
plt.suptitle('Muscle properties', y=1, fontsize=16)
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```
